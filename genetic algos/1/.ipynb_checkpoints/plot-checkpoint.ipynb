```python
import matplotlib.pyplot as plt
```

```python
with open("1.txt", "r") as file:
    lines = file.readlines()
for i, line in enumerate(lines):
    lines[i] = line.split(", ")[1].strip()
thread1 = list(map(float, lines))
print(thread1)
```

```python
with open("6.txt", "r") as file:
    lines = file.readlines()
for i, line in enumerate(lines):
    lines[i] = line.split(", ")[1].strip()
thread6 = list(map(float, lines))
print(thread6)
```

```python
with open("12.txt", "r") as file:
    lines = file.readlines()
for i, line in enumerate(lines):
    lines[i] = line.split(", ")[1].strip()
thread12 = list(map(float, lines))
print(thread12)
```

```python
size = list(range(100, 1001, 100))
print(size)
```

```python
plt.plot(size, thread1)
plt.plot(size, thread6)
plt.plot(size, thread12)
plt.legend(["1 thread", "6 threads", "12 threads"])
plt.title("Time cost without intrinsics")
plt.xlabel("Matrix size")
plt.ylabel("Time in seconds")
plt.show()
```

```python
with open("optimized_1.txt", "r") as file:
    lines = file.readlines()
for i, line in enumerate(lines):
    lines[i] = line.split(", ")[1].strip()
optimized_thread1 = list(map(float, lines))
```

```python
plt.plot(size, thread1)
plt.plot(size, optimized_thread1)
plt.legend(["simple", "optimized"])
plt.title("Time cost with and without O3 optimizations (1 thread)")
plt.xlabel("Matrix size")
plt.ylabel("Time in seconds")
plt.show()
```
