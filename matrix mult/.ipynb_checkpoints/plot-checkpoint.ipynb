```python
import matplotlib.pyplot as plt
```

```python
core1 = []
with open("1.txt", "r") as file:
    lines = file.readlines()
for i, line in enumerate(lines):
    lines[i] = line.split(", ")[1].strip()
core1 = list(map(float, lines))
print(core1)
```

```python
core6 = []
with open("6.txt", "r") as file:
    lines = file.readlines()
for i, line in enumerate(lines):
    lines[i] = line.split(", ")[1].strip()
core6 = list(map(float, lines))
print(core6)
```

```python
core12 = []
with open("12.txt", "r") as file:
    lines = file.readlines()
for i, line in enumerate(lines):
    lines[i] = line.split(", ")[1].strip()
core12 = list(map(float, lines))
print(core12)
```

```python
size = list(range(100, 1001, 100))
print(size)
```

```python
plt.plot(size, core1)
plt.plot(size, core6)
plt.plot(size, core12)
plt.legend(["1 core", "6 cores", "12 cores"]) 
plt.show()
```
